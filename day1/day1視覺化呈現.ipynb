```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Noto Sans CJK TC']
%matplotlib inline
```

```python
sid2337= pd.read_csv('旺宏.csv',index_col=0)['2019':]
sid2344= pd.read_csv('華邦電.csv',index_col=0)['2019':]
df= pd.DataFrame({'旺宏':sid2337['close'],'華邦電':sid2344['close']})
_= df.plot(figsize=(12,8))
```

```python
_= plt.hist((df['旺宏'].pct_change().dropna())*100, bins=20)
```
